```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df =pd.read_csv("StudentsPerformance.csv") # load the data using pandas
df.head() # 
```

### scatter plot using relplot

```python
sns.relplot(x='reading score',y='writing score',data=df,hue='gender')
# comparing the relation between reading and writing score of different gender 
# observation :- positive relationship
```

### line plot using relplot

```python
sns.relplot(x='reading score',y='writing score',data=df,hue='gender',col='gender',kind='line')
# comparing the relation between reading and writing score of different gender separately
# observation :- both show positve relatonship
```

### strip plot using catplot

```python
sns.catplot(x='race/ethnicity',y='math score',data=df,kind='strip')
# comparing the math score for different race and ethinicity
# observation:- race GROUP E has highest maths score
```

### pie plot

```python
df['parental level of education'].value_counts()
```

```python
a=list(df['parental level of education'].value_counts().keys())
a
```

```python
b=list(df['parental level of education'].value_counts().values)
b
```

```python

sns.set(rc={'figure.figsize':(5,5)})
plt.pie(b,labels=a,wedgeprops={'ec':'k'},autopct='%1.1f%%',shadow=True) # ec edgecolor explode autopct to show %
plt.title('parental level of education',color='k')
plt.show() 


# comparing parental level of education
# observaton :- very few has done master degree and most of them is graduated from some college
```

### barplot using catplot

```python
sns.catplot(x='parental level of education',y='math score',data=df,kind='bar',hue='gender')
plt.xticks(rotation = 60)
# comapring different gender ,parental level of education and maths score
```

### boxplot 

```python
sns.boxplot(df['writing score'])
plt.show()
# drawing boxplot of writing score
# observation :- some score is less than 22 and range is btw 60 to 80 
```

### by grouping

```python
a = df.groupby("parental level of education")['writing score'].sum()
a
```

```python
plt.bar(a.keys(),a.values)
plt.xticks(rotation = 60)
# grouping parental level of education
# apllying sum
# observation :- parents with associate's degree has score highest in writing
```
