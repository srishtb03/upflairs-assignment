#### 3. Make a group on “processor_brand” and find out the minimum and maximum price of all the processor_brand laptop.

```python
group=df.groupby("processor_brand") # group on "processor_brand"
group
```

```python
df.keys()
```

```python
group["Price"].agg(min_pr='min',max_pr='max') # the minimum and maximum price of 
# all the processor_brand laptop.
```
